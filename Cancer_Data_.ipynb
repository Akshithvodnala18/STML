```python
import pandas as pd

# Specify the engine explicitly to handle various Excel formats.
# Use 'openpyxl' for .xlsx files and 'xlrd' for older .xls files.
try:
    data = pd.read_excel('/content/drive/MyDrive/breast_cancer_survival.csv', engine='openpyxl')  # Try openpyxl first
except Exception as e:
    try:
        data = pd.read_excel('/content/drive/MyDrive/breast_cancer_survival.csv', engine='xlrd')  # Try xlrd if openpyxl fails
    except Exception as e:
        #If xlrd also fails, file might be CSV
        try:
            data = pd.read_csv('/content/drive/MyDrive/breast_cancer_survival.csv')
        except Exception as e:
            print(f"Error reading the file: {e}")

data.head(50)
```

```python
data.dtypes
```

```python
y = data['Age']  # Example: Accessing by a column name that exists in the dataset
x = data.drop(['Age'], axis=1)
print(x)
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
categorical_columns = ['Utility_Name','Utility_State','Utility_Type']
for column in categorical_columns:
    # Check if column is present in the dataframe before processing
    if column in x.columns:
        x[column] = labelencoder.fit_transform(x[column])
    else:
        print(f"Column '{column}' not found in the DataFrame.")

print(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print("Training Features:", x_train)
print("Training Target:", y_train)
print("Testing Features:", x_test)
print("Testing Target:", y_test)
```

```python
# Convert columns to numeric if they contain numeric strings
for column in x_train.columns:
    # Try converting to numeric, errors='coerce' will give NaN for non-numeric
    x_train[column] = pd.to_numeric(x_train[column], errors='coerce')
    x_test[column] = pd.to_numeric(x_test[column], errors='coerce')

# Recalculate min and max
x_min = x_train.min()
x_max = x_train.max()

# Perform min-max scaling
x_train = (x_train - x_min) / (x_max - x_min)
x_test = (x_test - x_min) / (x_max - x_min)

# Fill NaN values (if any introduced by to_numeric)
x_train.fillna(0, inplace=True)
x_test.fillna(0, inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
