Question 2:

Find log-likelihood values for 50, 75, and 80. For California_housing [‘latitude’] and find the Max likelihood values.

```python
mu = np.mean(d['latitude'])

sigma = np.var(d['latitude'], ddof=1)

def log_likelihood(x, mu,sigma):
  return -0.5*(np.log(2*np.pi*sigma**2)+ ((x-mu) **2/sigma**2))


  values = [50, 75, 88]

  log_likelihoods= [(log_likelihood (v, mu, sigma) for v in values)]

  print("Mean", mu )

  print("Variance:", sigma)

  print("Log-Likelihoods for values [50, 75, 88]:", log_likelihoods)
```

Question 3:

```python
import matplotlib.pyplot as plt
columns = ['latitude', 'total_rooms', 'housing_median_age', 'total_bedrooms']

bins = 5

for column in columns:
  plt.figure(figsize=(10, 6))
  plt.hist(d[column], bins=bins, density=True, alpha=0.6, color='blue', edgecolor="black") #Removed extra indent
  plt.title(f'Density Plot for {column}') #Removed extra indent
  plt.xlabel(column) #Removed extra indent
  plt.ylabel('Density') #Removed extra indent
  plt.grid(True)
  plt.show()
```
